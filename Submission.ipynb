```python
# To change the type of data and converting the data to the normal form

df['binnedInc']=df['binnedInc'].str.replace('(','')
df['binnedInc']=df['binnedInc'].str.replace('[','')
df['binnedInc']=df['binnedInc'].str.replace(']','')
```

```python
# Converting the data type of the particluar column 'binnedInc' to float

x=df['binnedInc'].str.split(',',expand=True).astype(float)
```

```python
# Viewing the data according to have the data types of every column in a correct way.

x=df['binnedInc'].str.split(',',expand=True).astype(float)
y=(x[0]+x[1])/2
df['binnedInc']=y
df.head()
```

```python
# Statistical Summary of the dataframe

df.describe()
```

```python
# Checking the outliers with in the data

for i in df:
    if (i=='Geography'):
        continue
    else:
        plt.figure()
        df.boxplot(column=[i])
```

- Observation : 

 All the variables including the dependent variable 'target_deathrate' has outliers.

```python
# Count of Outliers with in the Target Variable below Lower Whisker

print('Count of outliers below lower whisker is :',(df['TARGET_deathRate']<df['TARGET_deathRate'].quantile(0.25)-(1.5*(st.iqr(df['TARGET_deathRate'])))).sum())
```

```python
# Count of Outliers with in the Target Variable above Upper Whisker

print('Count of outliers above upper whisker is :',(df['TARGET_deathRate']>df['TARGET_deathRate'].quantile(0.75)+(1.5*(st.iqr(df['TARGET_deathRate'])))).sum())
```

```python
# Dropping Outliers from the dependent variable

# Since the target variable has outliers less then 10% of the data, drop the outliers

df1=df[(df['TARGET_deathRate']>df['TARGET_deathRate'].quantile(0.25)-(1.5*(st.iqr(df['TARGET_deathRate']))))&(df['TARGET_deathRate']<df['TARGET_deathRate'].quantile(0.75)+(1.5*(st.iqr(df['TARGET_deathRate']))))]
```

```python
# Checking for the shape after dropping outliers with in the target variable

df1.shape
```

```python
# Checking for null values

df1.isna().sum()
```

```python
# From the above line of code  'PctEmployed16_Over' have missing value less then 10% it is imputed with median

df1['PctEmployed16_Over']=df1['PctEmployed16_Over'].fillna(df1['PctEmployed16_Over'].median())
```

```python
# Checking for the distribution of features

for i in df1.columns:
    if (i=='Geography' or i=='PctSomeCol18_24' or i=='PctPrivateCoverageAlone'):
        continue
    else:
        plt.figure()
        sns.distplot(df1[i],kde=False,color='g',bins=10,rug=True)
```

```python
# Checking for the correlation between features

df1.corr()
```

```python
# Plotting a heatmap correspondingly

plt.figure(figsize=(30,15))
sns.heatmap(df1.corr(),annot=True)
```

- Observation:

 From the heatmap it becomes clear that there lies a multicollinarity between the variables

```python
# Performing OLS test to check the significant features

# Importing warnings 
import warnings 
warnings.filterwarnings('ignore')
# Importing stats required for OLS
import statsmodels.api as sm
# Dropping the 'geography' ,'pctsomecol18_24' ,'pctprivatecoveragealone' as they have the most missing values.
# Assinging the x and y variables 
x = df1.drop(['Geography' ,'PctSomeCol18_24','PctPrivateCoverageAlone','TARGET_deathRate'],axis=1)
x_constant = sm.add_constant(x)
y = df1['TARGET_deathRate']
y1 = list(y)
```

```python
# Fitting the model & Summary of the OLS test

model = sm.OLS(y1,x_constant).fit()
model.summary()
```

- Observation :

    Durbin-Watson score is 1.789 which indicates there lies no autocorrelation between the dependent features

```python
# Multicollinearity with VIF

from statsmodels.stats.outliers_influence import variance_inflation_factor
vif = [variance_inflation_factor(x_constant.values, i) for i in range(x_constant.shape[1])]
pd.DataFrame({'vif': vif[1:]}, index=x.columns)
```

```python
# Finding the features which has high multicollinearity

vif_=pd.DataFrame({'vif': vif[1:]}, index=x.columns)
vif_[vif_['vif']>10]
```

```python
# Features after the Statisticak test

features=[]
a=model.pvalues
for i in range(a.shape[0]):
    if  a[i]<0.05:
        features.append(a.index[i])
    else:
        continue
print(features)
```

```python
X = df1[[ 'avgAnnCount', 'avgDeathsPerYear', 'incidenceRate', 'popEst2015', 'MedianAgeMale', 'PercentMarried', 'PctHS18_24', 'PctHS25_Over', 'PctBachDeg25_Over', 'PctEmployed16_Over', 'PctPrivateCoverage', 'PctEmpPrivCoverage', 'PctOtherRace', 'PctMarriedHouseholds', 'BirthRate']]
```

```python
# Checking the relationship of depedent feature using scatter plot

for i in X.columns:
    plt.figure()
    sns.scatterplot(x=df1['TARGET_deathRate'],y=i,data=df1)
```

```python
# Checking the distribution of the data with in the features

for i in X.columns:
        plt.figure()
        sns.kdeplot(X[i])
```

```python
# Checking for the features which are not normally distributed

nor=[]
for i in  X.columns:
    if st.shapiro(X[i])[1]<0.05:
        nor.append(i)
    else:
        continue
print(nor)
```

```python
A = X[['avgAnnCount', 'avgDeathsPerYear', 'incidenceRate', 'popEst2015', 'MedianAgeMale', 'PercentMarried', 'PctHS18_24', 'PctHS25_Over', 'PctBachDeg25_Over', 'PctEmployed16_Over', 'PctPrivateCoverage', 'PctEmpPrivCoverage', 'PctOtherRace', 'PctMarriedHouseholds', 'BirthRate']]
```

```python
# Collecting the right/left skewed features

right=[]
left=[]
for i in  A.columns:
    if st.skew(A[i])>0.5:
        right.append(i)
    elif st.skew(A[i])<-0.5:
            left.append(i)
    else:
        continue
print('Right Skewed Features :\n ', right,'\n\nLeft Skewed Features :\n ',left)
```

```python
# Transformation

A['avgDeathsPerYear']=np.log((A['avgDeathsPerYear']))
A['avgAnnCount']=np.log((A['avgAnnCount']))
A['popEst2015']=np.log((A['popEst2015']))
A['PctBachDeg25_Over']=np.log((A['PctBachDeg25_Over']))
A['PctOtherRace']=(np.log((A['PctOtherRace'])+1))
A['BirthRate']=np.sqrt((A['BirthRate']))
A['PercentMarried']=((A['PercentMarried'])**2)
A['PctMarriedHouseholds']=((A['PctMarriedHouseholds'])**2)
```

```python
# Checking for skewness after transformation

for i in A:
    a=st.skew(A[i])
    print(i,':  ',a)
```

```python
# Distribution after transformation

for i in A.columns:
        plt.figure()
        sns.kdeplot(A[i])
```

- Observation:
    
    All the graph shows the reduction in the skewness and looks like most of the features got normally distributed except 'PctOtherRace'.

```python
# Model Building

# Train Test Split

from sklearn.model_selection import train_test_split
X_train, X_test , y_train, y_test = train_test_split(A,y1, test_size = 0.30, random_state = 20)
```

```python
# Initializing the Linear Regression Algorithm & Fitting the data

from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(X_train, y_train)
```

```python
# Checking for r2 score for the train data

print('r2 Score for Train Data :',lin_reg.score(X_train, y_train))
```

```python
# Checking for r2 score for the train data

print('r2 Score for Test Data :',lin_reg.score(X_test, y_test))
```

```python
# Prediction on test data

y_predict = lin_reg.predict(X_test)
```

```python
# Rmse Score of the model

from sklearn.metrics import mean_squared_error as ms
print('Rmse Score :',np.sqrt(ms(y_predict,y_test)))
```

```python
# Visualizing the predicted data

sns.scatterplot(y_predict,y_test)
plt.xlabel('y_predict')
plt.ylabel('y_test')
plt.show()
```

```python
# Visualization of distribution of the predicted data  

a=y_predict-y_test
sns.kdeplot(np.array(a))
plt.xlabel('y_predict-y_test')
plt.show()
```

```python
# Implementing the quadratic equation

from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree = 2) 
X_poly = poly.fit_transform(A) 
  
X_train, X_test , y_train, y_test = train_test_split(X_poly,y1, test_size = 0.30, random_state = 207)
```

```python
# Initializing & fitting the data

lin2 = LinearRegression() 
lin2.fit(X_train,y_train) 
```

```python
# r2 Score for the Train Data

print('r2 Score for Train Data :',lin2.score(X_train, y_train))
```

```python
# r2 Score for the Test Data

print('r2 score for Test Data :',lin2.score(X_test, y_test))
```

```python
# Prediction on Test Data

y_predict=lin2.predict(X_test)
```

```python
#  Rmse Score

print('Rmse Score :',np.sqrt(ms(y_predict,y_test)))
```

```python
#  Visualizing the Predicted Data

sns.scatterplot(y_predict,y_test)
plt.xlabel('y_predict')
plt.ylabel('y_test')
plt.show()
```

```python
# Visualization of distribution of the predicted data  

a=y_predict-y_test
sns.kdeplot(np.array(a))
plt.xlabel('y_predict-y_test')
plt.show()
```

```python
# Cross Validation Score

from sklearn.model_selection import cross_val_score
scores1 = cross_val_score(lin2,X=X_train,y=y_train, cv=10)
print ('Cross-validated scores:', scores1)
```

```python
# Getting the Mean & Standard Deviation of the data

print('Score_mean',scores1.mean(),': Score_std',scores1.std())
```
